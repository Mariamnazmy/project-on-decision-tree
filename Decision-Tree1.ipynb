```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('dataa.csv')
```

```python
data.head(6)
```

```python
data.describe()
```

```python
data1=data.replace(np.nan,data.mean())
data1.head()
```

```python
data1.describe()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['Sex'] = le.fit_transform(data['Sex'])
data['Race'] = le.fit_transform(data['Race'])
data['Marital status?'] = le.fit_transform(data['Marital status?'])
data['Education'] = le.fit_transform(data['Education'])
data['Incomes'] = le.fit_transform(data['Incomes'])
data['Employement'] = le.fit_transform(data['Employement'])
data['Read any printed books during last 12months?'] = le.fit_transform(data['Read any printed books during last 12months?'])
data['Read any audiobooks during last 12months?'] = le.fit_transform(data['Read any audiobooks during last 12months?'])
data['Read any e-books during last 12months?'] = le.fit_transform(data['Read any e-books during last 12months?'])
data['Last book you read'] = le.fit_transform(data['Last book you read'])
data['Do you happen to read any daily news or newspapers?'] = le.fit_transform(data['Do you happen to read any daily news or newspapers?'])
data['Do you happen to read any magazines or journals?'] = le.fit_transform(data['Do you happen to read any magazines or journals?'])

data.head(7)
```

```python
data2=(data-data.min())/(data.max()-data.min(0))
data2.head()
```

```python
features = data2.loc[:, 'Age': 'Do you happen to read any daily news or newspapers?']
features
```

```python
label = data2['Do you happen to read any magazines or journals?']
```

```python
label
```

```python
y_Actual=np.asarray(label)
y_Actual
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
res = model.fit(features,label)
```

```python
from sklearn import metrics
```

```python
y_predcit = model.predict(features)
```

```python
y_predcit
```

```python
data2.head()
```

```python
print('Accuracy = ', metrics.accuracy_score(label, y_predcit))
```

```python
from sklearn import tree
```

```python
tree.plot_tree(model, feature_names=['Age', 'Sex', 'Race', 'Marital status?', 'Education','Incomes', 'Employement'
                                     ,'book you read last 12m?', 'Read any printed books during last 12months?','Read any audiobooks during last 12months?','Read any e-books during last 12months?','Last book you read','Do you happen to read any daily news or newspapers?'], class_names=True)
```

```python
data_predict=y_predcit
data_actual=y_Actual
dataf={'y_Actual': data_actual,
     'y_predcit':data_predict }

df = pd.DataFrame(dataf,columns=['y_Actual','y_predcit'])
df.head()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
confusion_matrix = pd.crosstab(df['y_Actual'], df['y_predcit'], rownames=['Actual'], colnames=['Predicted'])
print (confusion_matrix)
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
sn.heatmap(confusion_matrix, annot=True)
plt.show()
```
